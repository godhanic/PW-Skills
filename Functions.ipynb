# Functions


1. What is the difference between a function and a method in Python?

*  A function is a block of reusable code defined using `def`, and can exist independently.
A method is a function that is associated with an object (usually a class instance).
Functions are called like `print("hello")`, while methods are called on objects like `"hello".upper()`.
All methods are functions, but not all functions are methods.
Example: `len([1, 2, 3])` is a function; `[1, 2, 3].append(4)` is a method.




2.   Explain the concept of function arguments and parameters in Python.

*   Parameters are variables listed in a function definition; arguments are the actual values passed to the function when it's called.
Python supports different types of arguments: positional, keyword, default, and variable-length.
Parameters act as placeholders; arguments provide real data.
They allow functions to be flexible and reusable.
Example:

def greet(name):  # 'name' is a parameter
    print("Hello", name)

greet("Ajay")  # "Ajay" is an argument



3. What are the different ways to define and call a function in Python?

*   In Python, you can define a function using `def` or `lambda`.
Functions can be called directly by name, or assigned to variables and then called.
You can also pass functions as arguments or return them from other functions.
Anonymous functions use `lambda` for short operations.
Example:

def add(x, y): return x + y  
print(add(2, 3))        # Direct call  
func = add  
print(func(4, 5))          # Via variable  




4.  What is the purpose of the `return` statement in a Python function?

*    The `return` statement sends a result back to the caller from a function.
It also ends the function’s execution immediately.
If no `return` is used, the function returns `None` by default.
You can return any data type or even multiple values.
Example:

def square(x):
    return x * x

print(square(4))  # Output: 16


5.  What are iterators in Python and how do they differ from iterables?

*   An iterable is any object you can loop over (like lists, strings, tuples).
An iterator is an object with a `__next__()` method that returns items one at a time.
You get an iterator from an iterable using the `iter()` function.
Iterators are used for lazy evaluation and stop when `StopIteration` is raised.
Example:

nums = [1, 2, 3]
it = iter(nums)         # Iterator from iterable
print(next(it))         # Output: 1


6.  Explain the concept of generators in Python and how they are defined.


*   Generators are special functions that yield values one at a time using the `yield` keyword.
They allow lazy evaluation, saving memory for large data sets.
Each call to `next()` resumes the function from where it left off.
Generators are defined like normal functions but use `yield` instead of `return`.
Example:

def count_up():
    for i in range(3):
        yield i

gen = count_up()
print(next(gen))  # Output: 0



7. What are the advantages of using generators over regular functions?

*  Generators are memory-efficient as they yield items one at a time instead of storing them all at once.
They support lazy evaluation, making them ideal for large data or infinite sequences.
They pause and resume automatically, saving function state between calls.
Generators are simpler to write than using custom iterator classes.
Example:

def count_up():
    for i in range(1000000):
        yield i

This avoids creating a massive list in memory.


8. What is a lambda function in Python and when is it typically used?

*   A lambda function is an anonymous, one-line function defined using the `lambda` keyword.
It’s used for short, simple operations without needing a full `def` function.
Commonly used with functions like `map()`, `filter()`, and `sorted()`.
Lambda functions return the expression result automatically.
Example:

square = lambda x: x * x
print(square(5))  # Output: 25



 9.  Explain the purpose and usage of the `map()` function in Python.


*  The `map()` function applies a given function to all items in an iterable (like a list).
It returns a map object, which can be converted to a list or used in a loop.
It’s useful for transforming data without writing a loop.
Typically used with functions or lambda expressions.
Example:

nums = [1, 2, 3]
squares = list(map(lambda x: x**2, nums))
print(squares)  # Output: [1, 4, 9]


10.  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

*  `map()` applies a function to each item in an iterable and returns the result.
* `filter()` selects items from an iterable that meet a condition.
* `reduce()` (from `functools`) reduces all items to a single value using a function.
  They are often used with lambda for concise functional programming.
  Example:

from functools import reduce
nums = [1, 2, 3, 4]
print(list(map(lambda x: x*2, nums)))       # [2, 4, 6, 8]
print(list(filter(lambda x: x%2==0, nums))) # [2, 4]
print(reduce(lambda x, y: x + y, nums))     # 10

````python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

nums = [10, 3, 4, 7, 8]
print(sum_even_numbers(nums))

````

````python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True


for num in range(1, 201):
    if is_prime(num):
       print(num, "is a prime number")
````

````python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.a

# Example usage:
fib = FibonacciIterator(10)
for num in fib:
    print(num)
````

````python
# 6.  Write a generator function in Python that yields the powers of 2 up to a given exponent

def powers_of_two(max_exp):
    for i in range(max_exp + 1):
        yield 2 ** i

for val in powers_of_two(5):
    print(val)
````

````python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

data = [(1, 3), (4, 1), (2, 2), (5, 0)]
sorted_data = sorted(data, key=lambda x: x[1])

print(sorted_data)
````

````python
# 9.  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit


celsius = [0, 20, 37, 100]

fahrenheit = list(map(lambda c: c * 9/5 + 32, celsius))

print(fahrenheit)
````

````python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(text):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda ch: ch not in vowels, text))


input_str = "Hello World"
result = remove_vowels(input_str)
print(result)
````

````python
# 11. Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:

'''
Order Number   Book Title and Author                        Quantity   Price per Item
34587          Learning Python, Mark Lutz                   4          40.95
98762          Programming Python, Mark Lutz                5          56.80
77226          Head First Python, Paul Barry                3          32.95
88112          Einführung in Python3, Bernd Klein           3          24.99
```

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity.
The product should be increased by 10 € if the value of the order is smaller than 100.00 €.

Write a Python program using `lambda` and `map`.
'''



orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]

# Using lambda and map to process each order
final_orders = list(map(
    lambda order: (order[0], order[2] * order[3] if order[2] * order[3] >= 100 else order[2] * order[3] + 10),
    orders
))

print(final_orders)
````
