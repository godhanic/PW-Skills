1. Explain the differences between AI, ML, Deep Learning (DL), and Data Science (DS).

* Here’s a simple way to understand it:

  * **AI (Artificial Intelligence):** The broad idea of making machines “smart” to perform tasks like humans.
  * **ML (Machine Learning):** A subset of AI where machines learn from data to make predictions or decisions.
  * **DL (Deep Learning):** A subset of ML that uses neural networks with many layers to learn complex patterns.
  * **DS (Data Science):** The field of extracting insights from data using statistics, ML, and visualization—AI/ML can be part of it.



2. What are the types of machine learning? Describe each with one real-world example.

* following are the types of Machine Learning (ML):

  * **Supervised Learning**

    * **Description:** The model is trained on **labeled data** (input + correct output). It learns to map inputs to outputs.
    * **Example:** Email spam detection – the model learns from emails labeled “spam” or “not spam” to classify new emails.

  * **Unsupervised Learning**

    * **Description:** The model works with **unlabeled data** and finds hidden patterns or groupings.
    * **Example:** Customer segmentation – grouping customers based on purchasing behavior without predefined categories.

  * **Semi-Supervised Learning**

    * **Description:** Uses a **small amount of labeled data** and a large amount of unlabeled data to improve learning.
    * **Example:** Medical image analysis – few labeled X-rays with many unlabeled ones to detect diseases.

  * **Reinforcement Learning (RL)**

    * **Description:** The model learns by **trial and error**, receiving rewards or penalties for actions to maximize cumulative reward.
    * **Example:** Self-driving cars – learning to drive safely by getting feedback on actions (stay on road = reward, crash = penalty).

  * **Self-Supervised Learning** *(emerging type)*

    * **Description:** The model generates its own labels from the input data to learn representations.
    * **Example:** ChatGPT – predicts the next word in a sentence using massive text data without explicit labels.



3. Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.

 * **Overfitting**

    * **Definition:** The model learns the **training data too well**, including noise, so it performs poorly on new/unseen data.
    * **Analogy:** Memorizing answers instead of understanding concepts.
    * **Example:** A complex decision tree that perfectly classifies training data but fails on test data.

 * **Underfitting**

    * **Definition:** The model is **too simple** to capture the underlying pattern in the data, performing poorly on both training and test data.
    * **Analogy:** Trying to solve a complicated problem with a very basic method.
    * **Example:** A linear model trying to fit a highly nonlinear dataset.

 * **Bias-Variance Tradeoff**

    * **Definition:** The balance between:

      * **Bias:** Error from overly simple models (underfitting).
      * **Variance:** Error from overly complex models (overfitting).
    * **Goal:** Find the sweet spot where the model generalizes well to new data.



4. What are outliers in a dataset, and list three common techniques for handling them.


 * Data points that are **significantly different** from most of the data in a dataset.
 * They can **skew analysis**, affect models, or reveal interesting anomalies.
 * **Example:** In a dataset of human heights, a value of 250 cm would be an outlier.

 * **Three common techniques to handle outliers:**

    1. **Removal:** Simply remove the extreme values if they are errors or irrelevant.
    2. **Transformation:** Apply transformations like **log, square root, or scaling** to reduce their impact.
    3. **Imputation:** Replace outliers with **mean, median, or a percentile value**.

```python
# 8. Write a Python script to:
# ● Generate 1000 samples from a normal distribution.
# ● Introduce 50 random missing values.
# ● Fill missing values with the column mean.
# ● Plot a histogram before and after imputation


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)


# Step 1: Generate 1000 samples from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)
df = pd.DataFrame(data, columns=['Value'])

# Step 2: Introduce 50 random missing values
missing_indices = np.random.choice(df.index, size=50, replace=False)
df.loc[missing_indices, 'Value'] = np.nan

# Step 3: Plot histogram before imputation
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.hist(df['Value'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram before Imputation')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Step 4: Fill missing values with the column mean
df['Value'].fillna(df['Value'].mean(), inplace=True)

# Step 5: Plot histogram after imputation
plt.subplot(1,2,2)
plt.hist(df['Value'], bins=30, color='lightgreen', edgecolor='black')
plt.title('Histogram after Imputation')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

```

```python
# 10.
"""
You are working as a data scientist for a retail company. You receive a customer
transaction dataset that contains:
● Missing ages,
● Outliers in transaction amount,
● A highly imbalanced target (fraud vs. non-fraud),
● Categorical variables like payment method.
Explain the step-by-step data preparation plan you’d follow before training a machine learning
model. Include how you’d address missing data, outliers, imbalance, and encoding.

"""


import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from collections import Counter
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)




# Simple example dataset
data = {
    'Age': [25, 30, np.nan, 22, 40, np.nan],
    'TransactionAmount': [100, 2000, 150, 30000, 250, 120],
    'PaymentMethod': ['Card', 'Cash', 'Card', 'Card', 'Cash', 'Card'],
    'Fraud': [0, 0, 0, 1, 0, 1]
}

df = pd.DataFrame(data)

# 1. Handle missing values (Age) using median
df['Age'].fillna(df['Age'].median(), inplace=True)

# 2. Handle outliers in TransactionAmount using simple capping
df['TransactionAmount'] = df['TransactionAmount'].clip(lower=df['TransactionAmount'].quantile(0.05),
                                                       upper=df['TransactionAmount'].quantile(0.95))

# 3. Encode categorical variable PaymentMethod
df = pd.get_dummies(df, columns=['PaymentMethod'], drop_first=True)

# 4. Scale numerical features
scaler = MinMaxScaler()
df[['Age', 'TransactionAmount']] = scaler.fit_transform(df[['Age', 'TransactionAmount']])

# 5. Split features and target
X = df.drop('Fraud', axis=1)
y = df['Fraud']

# 6. Simple way to handle imbalance: oversample minority class manually
minority_class = df[df['Fraud'] == 1]
X = pd.concat([X, minority_class.drop('Fraud', axis=1)], ignore_index=True)
y = pd.concat([y, minority_class['Fraud']], ignore_index=True)

# 7. Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Show processed data
print("Processed Features:\n", X_train)
print("\nProcessed Target:\n", y_train)
print("\nClass distribution after oversampling:", Counter(y_train))


```
