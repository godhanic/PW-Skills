# Python OOPs Questions


1.   What is Object-Oriented Programming (OOP)?

*   Object-Oriented Programming (OOP) in Python is a programming paradigm based on the concept of "objects."
It allows you to model real-world entities using classesand objects.
OOP supports principles like encapsulation, inheritance, and polymorphism.
It helps in writing reusable, organized, and modular code.
In Python, everything is an object, and OOP is widely used for building scalable applications.


2.  What is a class in OOP ?

*   A class in Python is a blueprint for creating objects.
It defines attributes (variables) and methods (functions) that describe the behavior of the object.
Classes allow you to bundle data and functionality together.
You create an object by instantiating a class.
Use the `class` keyword to define a class in Python.



3. What is an object in OOP ?
*   An object is an instance of a class in OOP.
It represents a real-world entity with state (attributes) and behavior (methods).
Objects are created from classes using the class constructor.
Each object can have different values for its attributes.
In Python, everything (like strings, lists, functions) is treated as an object.



4.  What is the difference between abstraction and encapsulation ?

*   Abstraction hides complex implementation details and shows only essential features.
Encapsulation binds data and methods into a single unit (class) and restricts direct access.
Abstraction focuses on what an object does, not how it does it.
Encapsulation focuses on how to protect data from unauthorized access.
Both improve code modularity, but serve different purposes in design.


5. What are dunder methods in Python ?

*   Dunder methods in Python are special methods with double underscores before and after their names (like `__init__`, `__str__`).
They are also called magic methods and are used to customize class behavior.
For example, `__init__` initializes objects, and `__str__` defines how an object is printed.
They enable operator overloading and built-in functions to work with user-defined objects.
You don't call them directly; Python calls them automatically in specific situations.



6. Explain the concept of inheritance in OOP


*   Inheritance in OOP allows a class (child) to acquire properties and behaviors from another class (parent).
It promotes code reuse and establishes a relationship between classes.
The child class can use, override, or extend the functionality of the parent class.
In Python, inheritance is defined by passing the parent class to the child class.
It supports single, multiple, and multilevel inheritance.



7. What is polymorphism in OOP ?

*  Polymorphism in OOP means the ability of different classes to respond to the same method in different ways.
It allows functions or methods to process objects differently based on their class.
Common types are method overriding and method overloading.
It enhances flexibility and reusability in code.
In Python, polymorphism is achieved through inheritance and duck typing.



8. How is encapsulation achieved in Python ?

*   Encapsulation in Python is achieved by wrapping data (variables) and methods inside a class.
Access to data is controlled using access modifiers: public, protected (`_var`), and private (`__var`).
Private members can’t be accessed directly from outside the class.
Getter and setter methods are used to access or update private data.
This protects the internal state and enforces controlled access.



 9. What is a constructor in Python ?

*  A constructor in Python is a special method used to initialize objects when they are created.
It is defined using the `__init__()` method inside a class.
It sets the initial values of the object’s attributes.
The constructor is called automatically when an object is instantiated.
It helps in setting up necessary state or default values for the object.


10. What are class and static methods in Python ?

*  Class methods are defined with `@classmethod` and take `cls` as the first parameter; they can access and modify class state.
Static methods are defined with `@staticmethod` and don’t take `self` or `cls`; they behave like regular functions inside a class.
Class methods are used when you need to work with the class itself, not instances.
Static methods are used for utility functions related to the class.
Both can be called using the class name or an object.


11. What is method overloading in Python ?

* Method overloading in Python refers to using the same method name with different arguments.
Python doesn't support true method overloading like other languages.
The latest method definition overrides previous ones.
It can be mimicked using default parameters or `*args`/`**kwargs`.
This allows one method to handle different types or numbers of arguments.



12. What is method overriding in OOP ?

* Method overriding in OOP allows a subclass to provide a specific implementation of a method already defined in its superclass.
The method in the child class has the same name and parameters as in the parent class.
It is used to modify or extend the behavior of inherited methods.
Python automatically calls the overridden method from the child class.
This supports runtime polymorphism and flexibility in code design.



13. What is a property decorator in Python ?

*   The `@property` decorator in Python is used to define a method as a read-only property.
It allows you to access a method like an attribute, without using parentheses.
It is commonly used to encapsulate data and control access to instance variables.
You can also use `@<property>.setter` and `@<property>.deleter` to modify or delete the property.
This helps in clean, controlled access to class data.



14. Why is polymorphism important in OOP ?

* Polymorphism is important in OOP because it allows different classes to be treated as if they were the same type.
It enables code reusability and flexibility by allowing the same interface to work with different underlying forms.
It simplifies code by reducing the need for multiple condition checks.
Polymorphism supports extensibility, making it easier to add new features.
It enhances maintainability and scalability of object-oriented programs.



15. What is an abstract class in Python ?

*   An abstract class in Python is a class that cannot be instantiated directly.
It is used as a blueprint for other classes and may contain abstract methods (methods without implementation).
Abstract classes are defined using the `abc` module and `ABC` class.
You define abstract methods using the `@abstractmethod` decorator.
Subclasses must implement all abstract methods to become concrete (usable) classes.



16. What are the advantages of OOP ?

*  Object-Oriented Programming (OOP) offers better modularity by organizing code into classes and objects.
It promotes code reusability through inheritance.
OOP enhances data security with encapsulation.
It supports flexibility and scalability via polymorphism and abstraction.
Overall, OOP makes code easier to maintain, test, and extend.



17. What is the difference between a class variable and an instance variable ?

*  A class variable is shared across all instances of a class.
An instance variable is unique to each object and defined using `self`.
Class variables are defined outside any method, typically directly under the class.
Instance variables are usually set inside the `__init__()` method.
Changing a class variable affects all objects; changing an instance variable affects only that object.



18. What is multiple inheritance in Python ?

*   Multiple inheritance in Python allows a class to inherit from more than one parent class.
This lets the child class access attributes and methods from all its parent classes.
It is defined by passing multiple parent classes in parentheses.
Python uses the Method Resolution Order (MRO) to handle conflicts.
While powerful, it should be used carefully to avoid complexity.



19. Explain the purpose of ‘’__str__’ and ‘__repr__’ ‘ methods in Python.

*   The `__str__` method defines the human-readable string representation of an object (used by `print()`).
The `__repr__` method defines the developer-friendly representation (used in the interpreter).
If `__str__` is not defined, Python falls back to `__repr__`.
`__repr__` should ideally return a string that can recreate the object.
Both help in debugging and displaying objects in a readable format.



20. What is the significance of the ‘super()’ function in Python ?


*   The `super()` function in Python is used to call methods from a parent or superclass.
It allows you to access inherited methods without explicitly naming the parent class.
Commonly used in constructors to initialize the parent class.
It supports multiple inheritance by following Python’s Method Resolution Order (MRO).
`super()` makes code more maintainable and avoids duplication.


21.  What is the significance of the __del__ method in Python ?

*  The `__del__` method in Python is a destructor, called when an object is about to be destroyed.
It is used to perform cleanup tasks, like closing files or releasing resources.
Python’s garbage collector automatically calls `__del__` when an object’s reference count reaches zero.
It is not guaranteed to be called immediately or at all in some cases.
Use it cautiously, as improper use can lead to unexpected behavior or memory issues.


22. What is the difference between @staticmethod and @classmethod in Python?


*  `@staticmethod` defines a method that doesn't take `self` or `cls` and can't access class or instance data.
`@classmethod` takes `cls` as the first parameter and can access or modify class-level data.
Static methods are used for utility functions related to the class.
Class methods are often used for alternative constructors or class-specific behavior.
Both can be called using the class name or an instance.


23. How does polymorphism work in Python with inheritance ?

*  Polymorphism in Python with inheritance allows subclasses to override methods from a parent class.
You can use the same method name across different classes, and Python will call the appropriate version based on the object's class.
This enables writing generic code that works with different object types.
It supports runtime polymorphism, where method behavior is determined at execution time.
This makes code more flexible, reusable, and easier to maintain.


24. What is method chaining in Python OOP ?

*  Method chaining in Python OOP is a technique where multiple methods are called in a single line, one after another.
Each method returns `self`, allowing the next method to be called on the same object.
It helps write cleaner, more readable, and fluent code.
This is commonly used in builder patterns or data processing pipelines.
To enable it, each method in the chain must return the object (`return self`).



25. What is the purpose of the __call__ method in Python?

*  The `__call__` method in Python lets an object be called like a function.
It is defined inside a class to make its instances callable.
This allows you to add function-like behavior to objects.
It’s useful for things like decorators, function wrappers, or configurable objects.
Python calls `__call__()` automatically when the object is followed by parentheses.

```python
# 1. Create a parent class Animal with a method speak() that prints a generic message. Create a child class Dog that overrides the speak() method to print "Bark!".

# Parent class
class Animal:
    def speak(self):
        print("The animal makes a sound.")

# Child class
class Dog(Animal):
    def speak(self):
        print("Bark!")


a = Animal()
a.speak()

d = Dog()
d.speak()
```

```python
# 2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.

from abc import ABC, abstractmethod

# Abstract class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Circle class
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

# Rectangle class
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Example usage
c = Circle(5)
print("Circle area:", c.area())

r = Rectangle(4, 6)
print("Rectangle area:", r.area())
```

```python
# 3.  Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car and further derive a class ElectricCar that adds a battery attribute.

# Base class
class Vehicle:
    def __init__(self, vehicle_type):
        self.vehicle_type = vehicle_type

# Derived class from Vehicle
class Car(Vehicle):
    def __init__(self, vehicle_type, brand):
        super().__init__(vehicle_type)
        self.brand = brand

# Further derived class from Car
class ElectricCar(Car):
    def __init__(self, vehicle_type, brand, battery_capacity):
        super().__init__(vehicle_type, brand)
        self.battery_capacity = battery_capacity

    def display_info(self):
        print(f"Type: {self.vehicle_type}")
        print(f"Brand: {self.brand}")
        print(f"Battery: {self.battery_capacity} kWh")

# Example usage
e_car = ElectricCar("Four-Wheeler", "Tesla", 75)
e_car.display_info()
```

```python
# 4.  Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes Sparrow and Penguin that override the fly() method.

# Base class
class Bird:
    def fly(self):
        print("Some birds can fly.")

# Derived class 1
class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies high in the sky.")

# Derived class 2
class Penguin(Bird):
    def fly(self):
        print("Penguins cannot fly, they swim.")

# Polymorphic behavior
def bird_fly_test(bird):
    bird.fly()

# Example usage
b1 = Sparrow()
b2 = Penguin()

bird_fly_test(b1)
bird_fly_test(b2)

```

```python
# 5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.

class BankAccount:
    def __init__(self, initial_balance=0):
        self.__balance = initial_balance  # private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: ₹{amount}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn: ₹{amount}")
        else:
            print("Insufficient balance or invalid amount.")

    def check_balance(self):
        print(f"Current Balance: ₹{self.__balance}")

# Example usage
account = BankAccount(1000)
account.check_balance()
account.deposit(500)
account.withdraw(300)
account.check_balance()

```

```python
# 6. Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar and Piano that implement their own version of play().

# Base class
class Instrument:
    def play(self):
        print("Playing an instrument.")

# Derived class 1
class Guitar(Instrument):
    def play(self):
        print("Strumming the guitar.")

# Derived class 2
class Piano(Instrument):
    def play(self):
        print("Playing the piano.")

# Function to demonstrate polymorphism
def perform(instrument):
    instrument.play()

# Example usage
g = Guitar()
p = Piano()

perform(g)
perform(p)
```

```python
# 7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static method subtract_numbers() to subtract two numbers.

class MathOperations:
    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b

# Example usage
sum_result = MathOperations.add_numbers(10, 5)
print("Sum:", sum_result)

diff_result = MathOperations.subtract_numbers(10, 5)
print("Difference:", diff_result)
```

```python
# 8. Implement a class Person with a class method to count the total number of persons created

class Person:
    count = 0  # Class variable to track number of persons

    def __init__(self, name):
        self.name = name
        Person.count += 1  # Increment count whenever a new person is created

    @classmethod
    def get_person_count(cls):
        return cls.count

# Example usage
p1 = Person("Alice")
p2 = Person("Bob")
p3 = Person("Charlie")

print("Total Persons Created:", Person.get_person_count())
```

```python
# 9. Write a class Fraction with attributes numerator and denominator. Override the str method to display the fraction as "numerator/denominator".

class Fraction:
    def __init__(self, numerator, denominator):
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):
        return f"{self.numerator}/{self.denominator}"

# Example usage
f = Fraction(3, 4)
print(f)
```

```python
# 10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors.

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overloading the + operator
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    # For readable output
    def __str__(self):
        return f"({self.x}, {self.y})"

# Example usage
v1 = Vector(2, 3)
v2 = Vector(4, 5)
v3 = v1 + v2  # Calls v1.__add__(v2)

print("Vector 1:", v1)
print("Vector 2:", v2)
print("Sum:", v3)
```

```python
# 12.  Implement a class Student with attributes name and grades. Create a method average_grade() to compute the average of the grades.

class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades  # Should be a list of numbers

    def average_grade(self):
        if self.grades:
            return sum(self.grades) / len(self.grades)
        else:
            return 0

# Example usage
s = Student("Rahul", [85, 90, 78, 92])
print(f"{s.name}'s average grade: {s.average_grade():.2f}")
```

```python
# 14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.

# Base class
class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate

# Derived class
class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        base_salary = super().calculate_salary()
        return base_salary + self.bonus

# Example usage
emp = Employee("Ajay", 40, 200)
print(f"{emp.name}'s Salary: ₹{emp.calculate_salary()}")

mgr = Manager("Bijay", 40, 300, 5000)
print(f"{mgr.name}'s Salary: ₹{mgr.calculate_salary()}")
```

```python
# 15. Create a class Product with attributes name, price, and quantity. Implement a method total_price() that calculates the total price of the product.

class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def total_price(self):
        return self.price * self.quantity

# Example usage
p = Product("Laptop", 50000, 2)
print(f"Total price for {p.name}: ₹{p.total_price()}")

```

```python
# 16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.

from abc import ABC, abstractmethod

# Abstract base class
class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

# Derived class 1
class Cow(Animal):
    def sound(self):
        print("Cow says Moo")

# Derived class 2
class Sheep(Animal):
    def sound(self):
        print("Sheep says Baa")

# Example usage
c = Cow()
c.sound()

s = Sheep()
s.sound()
```

```python
# 17. Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that returns a formatted string with the book's details.

class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        return f"'{self.title}' by {self.author}, published in {self.year_published}."

# Example usage
book = Book("The Alchemist", "Paulo Coelho", 1988)
print(book.get_book_info())
```

```python
# 18. Create a class House with attributes address and price. Create a derived class Mansion that adds an attribute number_of_rooms.

# Base class
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

# Derived class
class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

    def display_info(self):
        print(f"Address: {self.address}")
        print(f"Price: ₹{self.price}")
        print(f"Number of Rooms: {self.number_of_rooms}")

# Example usage
m = Mansion("123 Luxury Lane", 50000000, 10)
m.display_info()
```
